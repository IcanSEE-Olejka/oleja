1 Лабараторная

```
import math as m


def z1(x):
  return m.sqrt(2*x+2*m.sqrt(x**2-4))/(m.sqrt(x**2-4)+x+2)

def z2(x):
  return 1/m.sqrt(x+2)

b = float(input('b = '))
if b >= 2:
  print(z1(b), z2(b))
else:
  print('Ошибка. Под корнем получается отрицательное число.')
```

2 Лабараторная

1 часть

```
import math as m

def line(k, b, x):
  return k*x+b

def circle(r, x):
  return -m.sqrt(r ** 2 - (x + 1) ** 2) - 0

def func(x):
  if x >= -5 and x < -3:
    return line(0, 1, x)
  elif x > -3 and x <= -1:
    return circle(2, x) 
  elif x > -1 and x <= 2:
    return line(0, -2, x)
  elif x > 2 and x <= 5:
    return line(1, -4, x) 
  else:
      return 'x в не диапозона. \nВведите x от -5 до 5.'
x = float(input('x='))
print(func(x))
```

2 часть

```
import math as m

def shot(r, x, y):
  first_quarter = y < m.sqrt(r ** 2 + (x - r) ** 2) - r and abs(x) < r and y > 0
  second_quarter = y > -m.sqrt(r ** 2 + (x - r) ** 2) - r and x > 0 and y > 0
  if first_quarter and second_quarter: 
    print(x, y)
    print('Попал!')
  else:
    print(x, y)
    print('Не попал!')
r = float(input('r = '))
x = float(input('x = '))
y = float(input('y = '))
shot(r, x, y)
```

3 Лабараторная 1 часть

```
def separate_line():
  print('+{:<4}+{:<4}+'.format('-' * 4, '-' * 4))

def head_table():
  separate_line()
  print('I{:<4}I{:<4}I'.format('x', 'y'))
  separate_line()

def body_table(x0, x, step):
  while x0 <= x:
    y = func(x0)
    print('I{:<4}I{:<4}I'.format(round(x0, 2), round(y, 2)))
    separate_line()
    x0 += step

body_table(-5, 5, 0.01)
```

```
def series():
  r = float(input('r = '))
  for i in range(10):
    print("Выстрел", i+1)
    x = float(input('x = '))
    y = float(input('y = '))
    shot(r, x, y)

series()
```

3 часть

```
import math
def taylor(x, n):
  summ = 0
  quotient = 0
  for i in range(n):
    quotient = ((x - 1) ** (i + 1))/((i + 1) * x ** (i + 1))
    summ += quotient
  return summ

x = float(input('x = '))
n = int(input('n = '))
if x <= 0.5:
  print('x должен быть больше, чем 0.5')
else:
  print(taylor(x, n))
```
